```python
import os
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.metrics import accuracy_score

def load_h5_model(model_path):
    """
    Load an H5 model.

    :param model_path: Path to the .h5 model file.
    :return: Loaded Keras model.
    """
    return load_model(model_path)

def preprocess_image(image_path, target_size):
    """
    Preprocess an image for the Keras model.

    :param image_path: Path to the image file.
    :param target_size: Target size for the model (height, width).
    :return: Preprocessed image as a NumPy array.
    """
    image = load_img(image_path, target_size=target_size)  # Resize and load image
    image = img_to_array(image)  # Convert to array
    image = image / 255.0  # Normalize to [0, 1]
    return np.expand_dims(image, axis=0)  # Add batch dimension

def predict_food(model, image_path, target_size):
    """
    Predict if an image contains food using the Keras model.

    :param model: Loaded Keras model.
    :param image_path: Path to the image file.
    :param target_size: Expected input size for the model (height, width).
    :return: Prediction (0 for No Food, 1 for Food).
    """
    # Preprocess the image
    image = preprocess_image(image_path, target_size)

    # Predict
    prediction = model.predict(image)
    return int(np.argmax(prediction))  # Assume output is a one-hot encoded vector

def evaluate_model(csv_file, folder_path, model_path):
    """
    Evaluate the Keras model against labelled data.

    :param csv_file: Path to the CSV file with ground truth labels.
    :param folder_path: Path to the folder containing images.
    :param model_path: Path to the .h5 model file.
    """
    # Load the CSV file
    data = pd.read_csv(csv_file)
    if 'File' not in data.columns or 'Food/No Food' not in data.columns:
        print("Error: The CSV file must contain 'File' and 'Food/No Food' columns.")
        return

    # Load the model
    model = load_h5_model(model_path)

    # Get target input size
    input_shape = model.input_shape[1:3]  # (height, width)

    # Perform predictions
    predictions = []
    ground_truth = []
    i = 0

    for _, row in data.iterrows():
        file_name = row['File']
        label = row['Food/No Food']
        image_path = os.path.join(folder_path, file_name)

        if not os.path.exists(image_path):
            print(f"Warning: File {file_name} not found in {folder_path}. Skipping.")
            continue

        prediction = predict_food(model, image_path, input_shape)
        predictions.append(prediction)
        ground_truth.append(int(label))

        print(f"{i} , {prediction} , {int(label)}")
        i+= 1

    # Compute accuracy
    accuracy = accuracy_score(ground_truth, predictions)
    print(f"Model Accuracy: {accuracy:.2%}")

# Example usage
csv_path = "/Users/onkar/Documents/comding/chomding/eating_detection/Eating Data/Aparajita - 04.10.23/annotations.csv"       # Path to the CSV file
folder_path = "/Users/onkar/Documents/comding/chomding/eating_detection/Eating Data/Aparajita - 04.10.23/photos/4thOctPhotosCopy"       # Path to the folder containing images
model_path = "/Users/onkar/Documents/comding/chomding/eating_detection/cnn_model.h5"   # Path to the .h5 model
evaluate_model(csv_path, folder_path, model_path)
```

```python
# for testing h5 model 

import os
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.metrics import accuracy_score, f1_score
from datetime import datetime

def load_h5_model(model_path):
    """
    Load an H5 model.
    """
    return load_model(model_path)

def preprocess_image(image_path, target_size):
    """
    Preprocess an image for the Keras model.
    """
    image = load_img(image_path, target_size=target_size)  # Resize and load image
    image = img_to_array(image)  # Convert to array
    image = image / 255.0  # Normalize to [0, 1]
    return np.expand_dims(image, axis=0)  # Add batch dimension

def predict_food(model, image_path, target_size):
    """
    Predict if an image contains food using the Keras model.
    """
    image = preprocess_image(image_path, target_size)
    prediction = model.predict(image)
    return int(np.argmax(prediction))  # Assume output is a one-hot encoded vector

def parse_timestamp(file_name):
    """
    Parse the timestamp from the filename and convert it to a human-readable format.
    """
    try:
        # Extract the timestamp part after the underscore
        timestamp_str = file_name.split('_')[1].replace('-', ':').split('.')[0]

        # Handle the format HH:MM:Milliseconds
        parts = timestamp_str.split(':')
        if len(parts) == 3:
            hours = int(parts[0])
            minutes = int(parts[1])
            milliseconds = int(parts[2])  # Extract milliseconds
            seconds = milliseconds // 1000  # Convert milliseconds to seconds
            microseconds = (milliseconds % 1000) * 1000  # Remaining milliseconds as microseconds

            # Create a datetime object
            timestamp = datetime(2023, 10, 4, hours, minutes, seconds, microseconds)  # Replace with actual date if needed
            return timestamp.strftime("%Y-%m-%d %H:%M:%S.%f")
        else:
            raise ValueError("Invalid timestamp format")
    except Exception as e:
        print(f"Error parsing timestamp for file {file_name}: {e}")
        return None


def evaluate_model(csv_file, folder_path, model_path, output_csv):
    """
    Evaluate the Keras model against labelled data.
    """
    # Load the CSV file
    data = pd.read_csv(csv_file)
    if 'File' not in data.columns or 'Food/No Food' not in data.columns:
        print("Error: The CSV file must contain 'File' and 'Food/No Food' columns.")
        return

    # Load the model
    model = load_h5_model(model_path)
    input_shape = model.input_shape[1:3]  # (height, width)

    predictions = []
    ground_truth = []
    result_data = []
    i = 0

    for _, row in data.iterrows():
        file_name = row['File']
        label = row['Food/No Food']
        image_path = os.path.join(folder_path, file_name)

        if not os.path.exists(image_path):
            print(f"Warning: File {file_name} not found in {folder_path}. Skipping.")
            continue

        # Predict the label
        prediction = predict_food(model, image_path, input_shape)
        predictions.append(prediction)
        ground_truth.append(int(label))

        # Parse timestamp
        timestamp = parse_timestamp(file_name)
        i += 1
        print(f"{i} , {int(label)} , {prediction} , {timestamp}")

        # Append to results
        result_data.append({
            "Real Value": int(label),
            "Predicted Value": prediction,
            "Timestamp": timestamp
        })

    # Compute accuracy and F1 score
    accuracy = accuracy_score(ground_truth, predictions)
    f1 = f1_score(ground_truth, predictions)
    print(f"Model Accuracy: {accuracy:.2%}")
    print(f"Model F1 Score: {f1:.2%}")

    # Write results to CSV
    results_df = pd.DataFrame(result_data)
    results_df.to_csv(output_csv, index=False)
    print(f"Results saved to {output_csv}")

# Example usage
csv_path = "/Users/onkar/Documents/comding/chomding/eating_detection/Eating Data/Murali - 17.11.23/annotations.csv"       # Path to the CSV file
folder_path = "/Users/onkar/Documents/comding/chomding/eating_detection/Eating Data/Murali - 17.11.23/murali_photos"       # Path to the folder containing images
model_path = "/Users/onkar/Documents/comding/chomding/eating_detection/cnn_model.h5"   # Path to the .h5 model
output_csv = "/Users/onkar/Documents/comding/chomding/eating_detection/Evaluation_Results.csv"  # Output CSV file
evaluate_model(csv_path, folder_path, model_path, output_csv)
```

```python
# for testing tflite model

import os
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.metrics import accuracy_score, f1_score
from datetime import datetime
from tensorflow.keras.preprocessing.image import load_img, img_to_array

def load_tflite_model(model_path):
    """
    Load a TFLite model using TensorFlow's Interpreter.
    """
    interpreter = tf.lite.Interpreter(model_path=model_path)
    interpreter.allocate_tensors()
    return interpreter

def preprocess_image(image_path, target_size):
    """
    Preprocess an image for the TFLite model.
    """
    image = load_img(image_path, target_size=target_size)  # Resize and load image
    image = img_to_array(image)  # Convert to array
    image = image / 255.0  # Normalize to [0, 1]
    return np.expand_dims(image, axis=0)  # Add batch dimension

def predict_food_tflite(interpreter, image_path, target_size):
    """
    Predict if an image contains food using the TFLite model.
    """
    # Preprocess the image
    image = preprocess_image(image_path, target_size)
    
    # Get input and output details
    input_details = interpreter.get_input_details()
    output_details = interpreter.get_output_details()
    
    # Set the input tensor
    interpreter.set_tensor(input_details[0]['index'], image)
    
    # Run inference
    interpreter.invoke()
    
    # Get the output tensor
    output_data = interpreter.get_tensor(output_details[0]['index'])
    return int(np.argmax(output_data))  # Assume output is a one-hot encoded vector

def parse_timestamp(file_name):
    """
    Parse the timestamp from the filename and convert it to a human-readable format.
    """
    try:
        # Extract the timestamp part after the underscore
        timestamp_str = file_name.split('_')[1].replace('-', ':').split('.')[0]

        # Handle the format HH:MM:Milliseconds
        parts = timestamp_str.split(':')
        if len(parts) == 3:
            hours = int(parts[0])
            minutes = int(parts[1])
            milliseconds = int(parts[2])  # Extract milliseconds
            seconds = milliseconds // 1000  # Convert milliseconds to seconds
            microseconds = (milliseconds % 1000) * 1000  # Remaining milliseconds as microseconds

            # Create a datetime object
            timestamp = datetime(2023, 10, 4, hours, minutes, seconds, microseconds)  # Replace with actual date if needed
            return timestamp.strftime("%Y-%m-%d %H:%M:%S.%f")
        else:
            raise ValueError("Invalid timestamp format")
    except Exception as e:
        print(f"Error parsing timestamp for file {file_name}: {e}")
        return None

def evaluate_tflite_model(csv_file, folder_path, model_path, output_csv):
    """
    Evaluate the TFLite model against labelled data.
    """
    # Load the CSV file
    data = pd.read_csv(csv_file)
    if 'File' not in data.columns or 'Food/No Food' not in data.columns:
        print("Error: The CSV file must contain 'File' and 'Food/No Food' columns.")
        return

    # Load the TFLite model
    interpreter = load_tflite_model(model_path)
    input_details = interpreter.get_input_details()
    input_shape = input_details[0]['shape'][1:3]  # (height, width)

    predictions = []
    ground_truth = []
    result_data = []
    i = 0

    for _, row in data.iterrows():
        file_name = row['File']
        label = row['Food/No Food']
        image_path = os.path.join(folder_path, file_name)
        if not os.path.exists(image_path):
            print(f"Warning: File {file_name} not found in {folder_path}. Skipping.")
            continue

        # Predict the label
        prediction = predict_food_tflite(interpreter, image_path, input_shape)
        predictions.append(prediction)
        ground_truth.append(int(label))

        # Parse timestamp
        timestamp = parse_timestamp(file_name)
        i += 1

        print(f"{i} , {int(label)} , {prediction} , {timestamp}")

        # Append to results
        result_data.append({
            "Real Value": int(label),
            "Predicted Value": prediction,
            "Timestamp": timestamp
        })

    # Compute accuracy and F1 score
    accuracy = accuracy_score(ground_truth, predictions)
    f1 = f1_score(ground_truth, predictions)
    print(f"Model Accuracy: {accuracy:.2%}")
    print(f"Model F1 Score: {f1:.2%}")

    # Write results to CSV
    results_df = pd.DataFrame(result_data)
    results_df.to_csv(output_csv, index=False)
    print(f"Results saved to {output_csv}")

# Example usage
csv_path = "/Users/onkar/Documents/comding/chomding/eating_detection/Eating Data/Murali - 17.11.23/annotations.csv"       # Path to the CSV file
folder_path = "/Users/onkar/Documents/comding/chomding/eating_detection/Eating Data/Murali - 17.11.23/murali_photos"       # Path to the folder containing images
model_path = "/Users/onkar/Documents/comding/chomding/eating_detection/quantized_cnn_model.tflite"  # Path to the .tflite model
output_csv = "/Users/onkar/Documents/comding/chomding/eating_detection/Evaluation_Results_TFLite2.csv"  # Output CSV file
evaluate_tflite_model(csv_path, folder_path, model_path, output_csv)
```

```python
import pandas as pd
import numpy as np

def process_predictions(csv1_path, csv2_path, output_csv_path):
    # Read the CSV files
    df1 = pd.read_csv(csv1_path)
    df2 = pd.read_csv(csv2_path)

    # Convert timestamps to pandas datetime for easier handling
    df1['timestamp'] = pd.to_datetime(df1['timestamp'])
    df2['timestamp'] = pd.to_datetime(df2['timestamp'])

    # Round timestamps to the nearest 5-second interval
    df1['rounded_time'] = df1['timestamp'].dt.floor('5S')
    df2['rounded_time'] = df2['timestamp'].dt.floor('5S')

    # Group by the 5-second intervals and calculate the mean predictions
    df1_avg = df1.groupby('rounded_time')['prediction'].mean().reset_index()
    df2_avg = df2.groupby('rounded_time')['prediction'].mean().reset_index()

    # Merge the two datasets on the rounded time
    merged = pd.merge(df1_avg, df2_avg, on='rounded_time', how='outer', suffixes=('_csv1', '_csv2'))

    # Take the average of the two predictions
    merged['final_prediction'] = merged[['prediction_csv1', 'prediction_csv2']].mean(axis=1)

    # Fill missing predictions with 0 if no data exists for a timestamp in one of the files
    merged = merged.fillna(0)

    # Save the results to a new CSV
    merged[['rounded_time', 'final_prediction']].to_csv(output_csv_path, index=False)

    print(f"Processed predictions saved to {output_csv_path}")

# Paths to input and output files
csv1_path = '/Users/onkar/Documents/comding/chomding/eating_detection/Evaluation_Results1.csv'
csv2_path = '/Users/onkar/Documents/comding/chomding/eating_detection/full_predictions (1).csv'
output_csv_path = 'final_predictions.csv'

# Call the function
process_predictions(csv1_path, csv2_path, output_csv_path)
```

```python
import os
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.metrics import accuracy_score, f1_score
from datetime import datetime
from tensorflow.keras.preprocessing.image import load_img, img_to_array

def load_tflite_model(model_path):
    """
    Load a TFLite model using TensorFlow's Interpreter.
    """
    interpreter = tf.lite.Interpreter(model_path=model_path)
    interpreter.allocate_tensors()
    return interpreter

def preprocess_image(image_path, target_size):
    """
    Preprocess an image for the TFLite model.
    """
    image = load_img(image_path, target_size=target_size)  # Resize and load image
    image = img_to_array(image)  # Convert to array
    image = image / 255.0  # Normalize to [0, 1]
    return np.expand_dims(image, axis=0)  # Add batch dimension

def predict_food_tflite(interpreter, image_path, target_size):
    """
    Predict if an image contains food using the TFLite model.
    """
    # Preprocess the image
    image = preprocess_image(image_path, target_size)
    
    # Get input and output details
    input_details = interpreter.get_input_details()
    output_details = interpreter.get_output_details()
    
    # Set the input tensor
    interpreter.set_tensor(input_details[0]['index'], image)
    
    # Run inference
    interpreter.invoke()
    
    # Get the output tensor
    output_data = interpreter.get_tensor(output_details[0]['index'])
    return int(np.argmax(output_data))  # Assume output is a one-hot encoded vector

def parse_timestamp(file_name):
    """
    Parse the timestamp from the filename and convert it to a human-readable format.
    """
    try:
        # Extract the timestamp part after the underscore
        timestamp_str = file_name.split('_')[1].replace('-', ':').split('.')[0]

        # Handle the format HH:MM:Milliseconds
        parts = timestamp_str.split(':')
        if len(parts) == 3:
            hours = int(parts[0])
            minutes = int(parts[1])
            milliseconds = int(parts[2])  # Extract milliseconds
            seconds = milliseconds // 1000  # Convert milliseconds to seconds
            microseconds = (milliseconds % 1000) * 1000  # Remaining milliseconds as microseconds

            # Create a datetime object
            timestamp = datetime(2023, 10, 4, hours, minutes, seconds, microseconds)  # Replace with actual date if needed
            return timestamp.strftime("%Y-%m-%d %H:%M:%S.%f")
        else:
            raise ValueError("Invalid timestamp format")
    except Exception as e:
        print(f"Error parsing timestamp for file {file_name}: {e}")
        return None

def evaluate_tflite_model(csv_file, folder_path, model_path, output_csv):
    """
    Evaluate the TFLite model against labelled data.
    """
    # Load the CSV file
    data = pd.read_csv(csv_file)
    if 'File' not in data.columns or 'Food/No Food' not in data.columns:
        print("Error: The CSV file must contain 'File' and 'Food/No Food' columns.")
        return

    # Load the TFLite model
    interpreter = load_tflite_model(model_path)
    input_details = interpreter.get_input_details()
    input_shape = input_details[0]['shape'][1:3]  # (height, width)

    predictions = []
    ground_truth = []
    result_data = []
    i = 0

    for _, row in data.iterrows():
        file_name = row['File']
        label = row['Food/No Food']
        image_path = os.path.join(folder_path, file_name)

        if not os.path.exists(image_path):
            print(f"Warning: File {file_name} not found in {folder_path}. Skipping.")
            continue

        # Predict the label
        prediction = predict_food_tflite(interpreter, image_path, input_shape)
        predictions.append(prediction)
        ground_truth.append(int(label))

        # Parse timestamp
        timestamp = parse_timestamp(file_name)
        i += 1
        print(f"{i} , {int(label)} , {prediction} , {timestamp}")

        # Append to results
        result_data.append({
            "Real Value": int(label),
            "Predicted Value": prediction,
            "Timestamp": timestamp
        })

    # Compute accuracy and F1 score
    accuracy = accuracy_score(ground_truth, predictions)
    f1 = f1_score(ground_truth, predictions)
    print(f"Model Accuracy: {accuracy:.2%}")
    print(f"Model F1 Score: {f1:.2%}")

    # Append results to the existing CSV file
    results_df = pd.DataFrame(result_data)
    file_exists = os.path.exists(output_csv)
    results_df.to_csv(output_csv, mode='a', index=False, header=not file_exists)
    print(f"Results appended to {output_csv}")

# Example usage
csv_path = "/Users/onkar/Documents/comding/chomding/eating_detection/Eating Data/Shantanu - 19.11.23/annotations.csv"       # Path to the CSV file
folder_path = "/Users/onkar/Documents/comding/chomding/eating_detection/Eating Data/Shantanu - 19.11.23/shantanu_photos/shantanu_photos"       # Path to the folder containing images
model_path = "/Users/onkar/Documents/comding/chomding/eating_detection/quantized_cnn_model.tflite"  # Path to the .tflite model
output_csv = "/Users/onkar/Documents/comding/chomding/eating_detection/Evaluation_Results_TFLite1.csv"  # Output CSV file
evaluate_tflite_model(csv_path, folder_path, model_path, output_csv)
```
